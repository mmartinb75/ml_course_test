```python
df_train = df_train.replace({9:np.nan})
```

```python
df_train.isna().sum()
```

```python
df_w_bool = df_train.copy()
df_w_bool[df_w_bool.columns[5:]] = df_w_bool[df_w_bool.columns[5:]].replace({0:'NO', 1:'SI', np.nan:'NO DISPONIBLE'})
```
