```python
# Vamos a empezar con el preprocesamiento de los datos:
# Seguiremos el siguiente orden:
# 1. Valores Null
# 2. Variables categóricas.
# 3. Transformación de variables a distribución normal.
# 4. Estandarización.
```

```python
import pandas as pd
import numpy as np

train_df = pd.read_csv("../data/preprocessed/train_data.csv")
```

```python
# 1. Valores Null
# como recordamos, a primera vista no había ningún null
train_df.isna().sum().sum()
```

```python
# Pero esto era debido a que se codificaba como 9 los valores null.
# Volvamos a ponerlos a null

train_df = train_df.replace({9:np.nan})
train_df.isna().sum()
```

```python
# Ahora vemos que tenemos "No de Bedrooms con solo 1 null y el resto de variables categóricas con muchos nulls"
# A partir de aquí en vez de sobreescribir sobre el dataset original, vamos a ir generando copias con diferentes
# features que luego combinaremos para lo que queramos.

# primero vamos a transformar las variables binarias a categóricas con tres categorías: true, false y NO disponible.
# num bedrooms al tener solo 1 null, podemos imputarlo como el valor más frequente con un simple imputer
from sklearn.impute import SimpleImputer

df_to_cat = train_df.copy()
mapping = {0: 'No', 1: 'Si', np.nan: 'NO DISPONIBLE'}
df_to_cat[df_to_cat.columns[5:]] = df_to_cat[df_to_cat.columns[5:]].replace(mapping)

imputer = SimpleImputer(strategy='median')
df_to_cat['No. of Bedrooms'] = imputer.fit_transform(df_to_cat[['No. of Bedrooms']])
```

```python
df_to_cat.isna().sum()
```

```python
# 2. El siguiente paso va a ser generar variables one-hot encoding para todas las categorías. 5- para arriba.
from sklearn.preprocessing import OneHotEncoder

# Select columns from index 5 onwards
columns_to_encode = df_to_cat.columns[5:]

# Initialize OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fit and transform the data
encoded_data = encoder.fit_transform(df_to_cat[columns_to_encode])

# Create a DataFrame with the encoded data
#encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(columns_to_encode))
```

```python
encoded_df = pd.DataFrame(data=encoded_data, columns=encoder.get_feature_names_out(columns_to_encode), index=df_to_cat.index)
```
